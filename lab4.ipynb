```python
X=df.drop('price_range',axis=1)
y=df['price_range']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn_pandas import gen_features
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import StandardScaler, OneHotEncoder
# Визначення списку числових і категоріальних ознак
numeric_features = ['battery_power',"mobile_wt", "px_height","px_width","ram"]
bin=['blue',"dual_sim","four_g","three_g",'touch_screen', "wifi"]
categorical_features = [col for col in df.columns.drop("price_range") if col not in numeric_features and col not in bin]

# Генерація функцій для мапера
numeric_transformations = gen_features(
    columns=[[col] for col in numeric_features],
    classes=[
        {'class': StandardScaler}
    ]
)

categorical_transformations = gen_features(
    columns=[[col] for col in categorical_features],
    classes=[OneHotEncoder]
)
categorical_b = gen_features(
    columns=[[col] for col in bin]
)

```

```python
from sklearn_pandas import DataFrameMapper
# Побудова мапера
preprocessor = DataFrameMapper(
    numeric_transformations + categorical_transformations + categorical_b,
    df_out=True, default=False
)

# Навчання мапера на тренувальній вибірці та трансформація даних
preprocessor.fit(X_train)
X_train = preprocessor.transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Створення моделі k-NN
knn1 = KNeighborsClassifier()

# Список гіперпараметрів для тюнінгу
param_grid = {
    'n_neighbors': [1, 3, 5, 7, 9, 11, 13, 14, 15, 17,19,21,23,25,27,29,30,35,60,100],
    'weights': ['uniform', 'distance'],
}

# Пошук оптимальних гіперпараметрів за допомогою GridSearchCV
grid_search = GridSearchCV(knn1, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
best_mod=grid_search.best_estimator_
# Виведення кращих параметрів та звіт про результати
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
from sklearn.metrics import classification_report
bestknn_pred=best_mod.predict(X_test)
print(classification_report(y_test,bestknn_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

class CustomConfusionMatrix:
    def __init__(self, y_test, bestknn_pred, labels=None):
        self.y_test = y_test
        self.bestknn_pred = bestknn_pred
        self.labels = labels


    def plot(self, normalize=None):
        conf_matrix = confusion_matrix(self.y_test, self.bestknn_pred, normalize=normalize)
        
        plt.figure(figsize=(4, 3))
        if normalize == 'true':
            disp_dt = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['class 0', 'class 1', 'class 2', 'class 3'])
            disp_dt.plot(cmap='Blues')
            plt.title("\nConfusion Matrix (Normalized by Recall)")
        elif normalize == 'pred':
            disp_dt = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['class 0', 'class 1', 'class 2', 'class 3'])
            disp_dt.plot(cmap='Blues')
            plt.title("\nConfusion Matrix (Normalized by Precision)")
        else:
            disp_dt = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['class 0', 'class 1', 'class 2', 'class 3'])
            disp_dt.plot(cmap='Blues')
            plt.title("\nConfusion Matrix")
        plt.show()


```

```python
# Приклад використання для SVC
conf_matrixknn = CustomConfusionMatrix(y_test, bestknn_pred)
conf_matrixknn.plot()
conf_matrixknn.plot(normalize='true')
conf_matrixknn.plot(normalize='pred')
```
